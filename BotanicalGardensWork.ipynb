```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
url = 'https://docs.google.com/spreadsheets/d/142CWqxECjRfv2G3tb7TvEdFYV5yEabWqWNlmHwvlmqQ/edit#gid=1443409278'
export_url = url.replace('/edit#gid=', '/export?format=csv&gid=')

biopark_df = pd.read_csv(export_url)
biopark_df
```

```python
biopark_df['Timestamp'] = pd.to_datetime(biopark_df['Timestamp'])
```

```python
temperature_columns = ['Timestamp', 'East Temperature', 'West Temperature', 'Central Temperature']
temperature_df = biopark_df[temperature_columns].copy()
```

```python
before_091723_df = biopark_df[(biopark_df['Timestamp'] >= '2023-08-16') & (biopark_df['Timestamp'] < '2023-09-17')]

before_091723_df = before_091723_df[(before_091723_df['East Temperature'] >= 10) & (before_091723_df['West Temperature'] >= 10) & (before_091723_df['Central Temperature'] >= 10)]

february_df = biopark_df[(biopark_df['Timestamp'] >= '2024-02-01') & (biopark_df['Timestamp'] < '2024-02-29')]
```

```python
# Plot before 9/17/23 with Plotly Express

custom_colors = ['#f8ff33', '#33FFFF', '#ff33ff']
fig = px.line(before_091723_df, x='Timestamp', y=['East Temperature', 'West Temperature', 'Central Temperature'],
              title='Temperature From 8/16/23 to 9/17/23', color_discrete_sequence=custom_colors)

fig.update_xaxes(title_text='Date', title_font=dict(family="Georgia", size=16, color="blue"))
fig.update_yaxes(title_text='Temperature (Â°F)', title_font=dict(family="Georgia", size=16, color="blue"))
fig.update_layout(title_font=dict(family="Georgia", size=24, color="blue"))
fig.update_layout(plot_bgcolor='black')
fig.show()
```

```python
fig = px.line(before_091723_df, x='Timestamp', y=['East Temperature', 'West Temperature', 'Central Temperature'],
              title='Temperature from 8/16/23 to 9/17/23')

fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Temperature (Â°F)')
fig.show()
```

```python
fig = px.line(before_091723_df, x='Timestamp', y=['East Humidity', 'West Humidity', 'Central Humidity'], title='Humidity from 8/16/23 to 9/17/23')

fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Humidity')
fig.show()
```

```python
# Transpose the DataFrame to have timestamps along the x-axis and locations along the y-axis
transposed_df = before_091723_df[['East Temperature', 'West Temperature', 'Central Temperature']].T

# Create a heatmap for temperature data
fig = px.imshow(transposed_df,
                x=before_091723_df['Timestamp'],
                y=['East Temperature', 'West Temperature', 'Central Temperature'],
                aspect='auto',
                labels={'x': 'Timestamp', 'y': 'Location', 'color': 'Temperature (Â°C)'},
                title='Temperature Heatmap 8/16-9/17')
fig.show()
```
